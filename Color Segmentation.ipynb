```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('candy.jpg')
#convert it to RGB to be able to display in matplotlib
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
img_hsv = cv2.cvtColor(img,cv2.COLOR_RGB2HSV)
plt.title("Candy Image")
plt.imshow(img)
plt.show()
```

```python
#create mask for orange color
l_orange = np.array([5,141,221])
h_orange = np.array([17,255,255])

mask_orange = cv2.inRange(img_hsv,l_orange,h_orange)
result = cv2.bitwise_and(img,img,mask=mask_orange)
plt.subplot(1,2,1)
plt.imshow(mask_orange,cmap='gray')
plt.subplot(1,2,2)
plt.imshow(result)
plt.show()

#Closing on the mask to remove noise
kernel= np.ones((30,30), np.uint8)
mask_orange = cv2.morphologyEx(mask_orange, cv2.MORPH_CLOSE, kernel)
plt.title("Mask for orange color:")
plt.imshow(mask_orange,'gray')
plt.show()

#Draw contours on the image
orange_count = 0
img2 = img.copy()
_, contours, _  = cv2.findContours(mask_orange, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
for c in contours:
    if cv2.contourArea(c) > 50000:
        cv2.drawContours(img2,[c],0,(201,14,27),30)
        orange_count += 1

        
plt.title("Orange candies detected:")
plt.imshow(img2)
plt.show()
print("Total number of orange candies: ", orange_count);
```
